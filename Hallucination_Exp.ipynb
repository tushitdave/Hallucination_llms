#### Importing necessary libraries

```python
import openai
from sentence_transformers import SentenceTransformer, util
import numpy as np
from selfcheckgpt.modeling_selfcheck import SelfCheckBERTScore, SelfCheckNLI
import pandas as pd
pd.set_option('display.max_colwidth', None)
import torch 
import spacy
import json
```

```python
if torch.backends.mps.is_available():
    mps_device = torch.device("mps")
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def roman_numeral(num):
    val = [
        1000, 900, 500, 400,
        100, 90, 50, 40,
        10, 9, 5, 4,
        1
    ]
    syb = [
        "M", "CM", "D", "CD",
        "C", "XC", "L", "XL",
        "X", "IX", "V", "IV",
        "I"
    ]
    roman_num = ''
    i = 0
    while num > 0:
        for _ in range(num // val[i]):
            roman_num += syb[i]
            num -= val[i]
        i += 1
    return roman_num
```

#### Generate text samples based on a given prompt, with additional instructional text indicating the purpose of each sample. The goal is to facilitate the generation of diverse and contextually relevant responses.

```python
def generate_samples_with_instruct(prompt):
    return [
        openai.Completion.create(
            engine="gpt-3.5-turbo-instruct",
            prompt=f"{prompt}\n\n**Sample {roman_numeral(i + 1)}:**",
            max_tokens=200,
            temperature=0.7
        ).choices[0].text.lstrip('\n') for i in range(3)
    ]

# Generate three additional samples. The function returns the main output and the list of three samples.
def get_output_and_samples_with_instruct(prompt):
    return openai.Completion.create(
        engine="gpt-3.5-turbo-instruct",
        prompt=f"{prompt}\n\n",
        max_tokens=100,
        temperature=0.7
    ).choices[0].text, generate_samples_with_instruct(prompt)
```

```python
prompt_example = "Write a short story about an astronaut who discovers a new planet."

output_text, sampled_passages = get_output_and_samples_with_instruct(prompt_example)
```

```python
print("Generated Output:")
print(output_text)

print("\nSampled Passages:")
for i, passage in enumerate(sampled_passages):
    if i == 0:
        print(f"**Sample {roman_numeral(i + 1)}:** {passage}") 
    else:
        print(f"**Sample {roman_numeral(i + 1)}:** {passage}")
```

```python
def calculate_cosine_similarity(output, sampled_passages):
    model = SentenceTransformer('all-MiniLM-L6-v2')
    sentence_embeddings = model.encode(output).reshape(1, -1)
    sample_embeddings = [model.encode(sample).reshape(1, -1) for sample in sampled_passages]
    cos_similarities = [util.pairwise_cos_sim(sentence_embeddings, sample_emb).item() for sample_emb in sample_embeddings]
    cos_sim_mean = np.mean(cos_similarities)

    return round(cos_sim_mean, 2)

cosine_similarity_mean = calculate_cosine_similarity(output_text, sampled_passages)
print(f"Mean Cosine Similarity: {cosine_similarity_mean}")
```

```python
def calculate_bertscore(output, sampled_passages):
    sentences = [sent.text.strip() for sent in nlp(output).sents]

    # Calculate BERT scores
    selfcheck_bertscore = SelfCheckBERTScore(rescale_with_baseline=True)
    sent_scores_bertscore = selfcheck_bertscore.predict(
        sentences=sentences,
        sampled_passages=sampled_passages,
    )

    sample_names = [f"Sample {roman_numeral(i + 1)}" for i in range(len(sampled_passages))]
    df = pd.DataFrame({
        'Sentence_Number': range(1, len(sent_scores_bertscore) + 1),
        'Sentence_Text': sentences,
        'Hallucination_Score': sent_scores_bertscore,
        'Sample_Name': [sample_names[i % len(sample_names)] for i in range(len(sent_scores_bertscore))]
    })

    return df
```

```python
bertscore_df = calculate_bertscore(output_text, sampled_passages)
```

```python
def calculate_nli_scores(output, sampled_passages):
    sentences = [sent.text.strip() for sent in nlp(output).sents]

    # Calculate NLI scores
    selfcheck_nli = SelfCheckNLI(device=mps_device)  # set device to 'cuda' if GPU is available
    sent_scores_nli = selfcheck_nli.predict(
        sentences=sentences,  # list of sentences
        sampled_passages=sampled_passages,  # list of sampled passages
    )
    # Calculate BERT scores
    selfcheck_bertscore = SelfCheckBERTScore(rescale_with_baseline=True)
    sent_scores_bertscore = selfcheck_bertscore.predict(
        sentences=sentences,
        sampled_passages=sampled_passages,
    )
    # Create DataFrame with scores, sentence text, and sample names
    sample_names = [f"Sample {roman_numeral(i + 1)}" for i in range(len(sampled_passages))]
    df = pd.DataFrame({
        'Sentence_Number': range(1, len(sent_scores_nli) + 1),
        'Sentence_Text': sentences,
        'Probability_of_Contradiction': sent_scores_nli,
        'Hallucination_Score': sent_scores_bertscore,
        'Sample_Name': [sample_names[i % len(sample_names)] for i in range(len(sent_scores_nli))]
    })

    return df
```

```python
nli_scores_df = calculate_nli_scores(output_text, sampled_passages)
nli_scores_df
```
